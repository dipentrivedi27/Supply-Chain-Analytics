### Revenue by Category

```python
category_sales = (
    df.groupby('Category Name')['Sales']
      .sum()
      .sort_values(ascending=False)
)

print(category_sales.head(10))
```

### Revenue by Market

```python
Market_sales = (df.groupby('Market')['Sales'].sum().sort_values(ascending = False))
print(Market_sales.head(10))
```

### Revenue by Country

```python
Country_sales = (df.groupby('Order Country')['Sales'].sum().sort_values(ascending = False))
print(Country_sales.head(10))
```

### Revenue by Top Products

```python
Top_products_sales = (df.groupby('Product Name')['Sales'].sum().sort_values(ascending = False))
print(Top_products_sales.head(10))
```

### Top Products by Quantity Sold

```python
Products_Quantity_sales = (df.groupby('Product Name')['Order Item Quantity'].sum().sort_values(ascending = False))
print(Products_Quantity_sales.head(10))
```

### Top Products by Profit

```python
total_profit = (df.groupby('Product Name')['Order Profit Per Order'].sum().sort_values(ascending = False))
print(total_profit.head(10))
```

## ==>Late Delivery Analysis

```python
# Delay Days Column

df['Delay_Days'] = (df['Days for shipping (real)'] - df['Days for shipment (scheduled)'])
```

```python
# Late Orders Count

late_orders = (df['Delay_Days'] > 0).sum()
print(late_orders)
```

```python
#  On-Time Orders Count

on_time_orders = (df['Delay_Days'] <= 0).sum()

print(on_time_orders)
```

```python
late_percentage = (
    (df['Delay_Days'] > 0).mean()
) * 100

print(round(late_percentage, 2))
```

```python
print("Total Profit:", round(df['Order Profit Per Order'].sum(), 2))
print("Average Profit Per Order:", round(df['Order Profit Per Order'].mean(), 2))
```
